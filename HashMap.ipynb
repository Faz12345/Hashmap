```python
class DoubleHashing: #class 
    delData = (None,None) #placeholder for deletion
    def __init__(self,size,num): #initialising object
        self.size = size
        self.num = num
        self.table = [None]*self.size

    def hash1(self,key): #1st hash func
        return key % self.size
    
    def hash2(self,key): #2nd hash func
        return self.num - (key % self.num)
    
    def insert(self,key,val): #insert method for indexing
        idx1 = self.hash1(key)
        if self.table[idx1] is None or self.table[idx1] == self.delData: #if condition meet then put key and val into taable
            self.table[idx1] = (key,val)
        else: #check for index val
            i = 1
            idx2 = self.hash2(key)
            while i < self.size: #iterating till i exceed size of table
                idx = (idx1 + i*idx2) % self.size
                if self.table[idx] is None or self.table[idx] == self.delData:
                    self.table[idx] = (key,val)
                    return
                i += 1
    
    def delete(self,key):
        idx1 = self.hash1(key)
        if self.table[idx1] is not None and self.table[idx1][0] == key: #if the key is found then delete the key of that index
            self.table[idx1] = self.delData
        else: #go to another index
            i = 1
            idx2 = self.hash2(key)
            while i < self.size: #iteratint untill condition met
                idx = (idx1 + i*idx2) % self.size
                if self.table[idx] is None: #if the index is None then return false
                    return False
                if self.table[idx][0] == key: #if the key is found then delete the key of that index
                    self.table[idx] = self.delData
                    return True
                i += 1
    
    def search(self,key):
        idx1 = self.hash1(key)
        if self.table[idx1] is not None and self.table[idx1][0] == key: #if the key is found then return key
            return self.table[idx1][1]
        else: #go to another index
            i = 1
            idx2 = self.hash2(key)
            while i < self.size: #iteratint untill condition met
                idx = (idx1 + i*idx2) % self.size
                if self.table[idx] is None: #if the index is None then return false
                    return False
                if self.table[idx][0] == key: #if the key is found then return key
                    return self.table[idx][1]
                i += 1
    def display(self): #method for displaying HashTable
        for i in range(self.size):
            if self.table[i] is None:
                print(f'{i}:[]')
            elif self.table[i] == self.delData:
                print(f'{i}:[Deleted]')
            else:
                print(f'{i}:{self.table[i]}')
l = DoubleHashing(10,7)
l.insert(10,'A')
l.insert(43,'B')
l.insert(63,'C')
l.insert(20,'D')
l.insert(56,'E')
l.insert(45,'F')
l.insert(55,'G')
l.display()
    
```
